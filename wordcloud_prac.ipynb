# Step 2

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud
wordcloud = WordCloud(width = 800, height = 800,
                      background_color ='white',
                      min_font_size = 10).generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize = (4, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

# Step Customizations

```python
custom_wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="gray").generate(text)
plt.imshow(custom_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
plt.savefig('wordcloud-2.png', format='png')
```

## Using RGB?

Refer to [RGB link](https://www.w3schools.com/colors/colors_rgb.asp)

```python
# Example: Converting RGB to Hexadecimal
r, g, b = 77, 252, 0  # color?
hex_color = f'#{r:02x}{g:02x}{b:02x}'

# Now, use the hex_color in the WordCloud generation
from wordcloud import WordCloud
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color=hex_color).generate(text)

# Assuming you have the rest of your matplotlib setup for displaying the word cloud
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

# Save the image in the current directory with the name 'wordcloud.png'
plt.show()
plt.savefig('wordcloud.png', format='png')
```
